
> 2.b  Provide the following statistics for query string “avatar movie” of top 50 videos sorted by relevance in the US region ( 7 marks )

> Output expected: video ID, title, no of views, no of likes,no of comments exported to CSV file






Reference link: https://developers.google.com/youtube/v3/docs/videos/list

```python
# Code Block for Task 2.b


import googleapiclient.discovery 
import pprint
import csv

#sending request 
request = youtube.search().list(part="snippet", q="avatar movie", type="video",
                                maxResults= 50, regionCode="US", order = "relevance")
response = request.execute()


# Created a list of video IDs from the search response
video_ids = []
for item in response["items"]:
    video_ids.append(item["id"]["videoId"])
    
    
# Get the statistics for each video ID using the videos.list method
request = youtube.videos().list(part="statistics,snippet", id=",".join(video_ids))
response = request.execute()

#print(video_ids)
#print(response)

# Created a list of dictionaries with the required statistics for each video
stats = []
for item in response["items"]:
    stats.append({
        "video ID": str(item["id"]),
        "title": str(item["snippet"]["title"]),
        "no of views": int(item["statistics"]["viewCount"]),
        "no of likes": int(item["statistics"]["likeCount"]),
        "no of comments": int(item["statistics"]["commentCount"])        
     }) 
    
# Checked the dictionary values by Printing stats
#print(stats)

# Export the statistics to a CSV file
with open("avatar_movie_stats.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=stats[0].keys())
    writer.writeheader()
    writer.writerows(stats)

print("Statistics exported to avatar_movie_stats.csv")

#using Panda for printing the output table 
import pandas as pd 
df = pd.read_csv('avatar_movie_stats.csv')
df.head(len(df))
```

> 3.a Sort the data 2.b  by top 10 comments in descending order and consider the video IDs and Titles of top 10 videos which have highest comments. (3mark)



```python
# Code Block for Task 3.a

import csv

# Reading the exported data from the CSV file obtained in 2.b
with open("avatar_movie_stats.csv", "r") as csvfile:
    reader = csv.DictReader(csvfile)
    stats = list(reader)
    

# Sorting the data by the number of comments in descending order 
stats.sort(key=lambda x: int(x["no of comments"]), reverse=True)

#  Taking only the top 10 videos
top_10 = stats[:10]

#print(top_10)

# using pandas dataframe for printing the output table 
df = pd.DataFrame(top_10)


df[["video ID", "title", "no of comments"]].head(len(df))

```


> 3.b Use a suitable method to retrieve comments of those top 10 videos from 3.a. For doing this, write a program to loop through each video id from 3.a and pass in the part parameter set to "snippet", to retrieve basic details about the comments. Execute this request and print the response using the pprint() method.
 - Note: pprint() will print out the response from the API in a more human-readable format.
- Reference link:  [link](https://developers.google.com/youtube/v3/docs )


> **Output expected** : Use the python library “ pprint “ to print the output of the program with the following properties  
etag, 
items, 
id , 
kind, 
snippet and snippet to have the text display field which represents the comment of videos.






```python
# Code Block for Task 3.b

import pprint


comment_data = [] 

# Looping through each video ID from the top_10 list
for video in top_10:
    # Get the video ID and title
    video_id = video["video ID"]
    video_title = video["title"]
    

    comment_data.append("Video ID : " + video_id)
    comment_data.append("Video Title : " + video_title)
    comment_data.append("Comments:")

    # Create a request to get comments for the video using the commentThreads.list() method
    request = youtube.commentThreads().list(part="snippet", videoId=video_id, maxResults=100)
    comment_response = request.execute()

    #pprint.pprint(comment_response)          #to prient entire comment_reponse with all tags
    #comment_data.append(comment_response)
    
    #filtering the comment_response to print only selected values 
    for item in comment_response["items"]:
        comment_data.append({
        "Comments etag": item["etag"],
        "Comments id": item["id"],
        "Comments kind": item["kind"],
        "Comments snippet-text": item["snippet"]["topLevelComment"]["snippet"]["textDisplay"],
         })
        
pprint.pprint(comment_data)
```

>3.d Write a function to get  the likes vs views ratio of the top 10 videos obtained in 3.a with the highest comments (3 marks)




```python
# Code Block for Task 3.d

import pprint

#pprint.pprint(top_10)
       
# Defining a function to get the likes vs views ratio for a video
def get_ratio(video):
    # Extract the number of likes and views from the video
    likes = int(video["no of likes"])
    views = int(video["no of views"])

    # Return the ratio as a float
    return likes / views


# Created a list of dictionaries with the required statistics for each video
comment_data = []
for item in top_10[:10]:   
    # Get the ratio by calling the function
    ratio = get_ratio(item)
    Ratio = f"{ratio:.3f}"    # reducing the floating to 3 decimal points 
    comment_data.append({
        "video ID": str(item["video ID"]),
        "Title": str(item["title"]),
        "Ratio": float(Ratio)       
     }) 

#using pandas dataframe to print output     
df = pd.DataFrame(comment_data)   

print("The Output Table is below: ")
print("*Ratio = no of likes / no of views ")
df.head(len(df))


```
